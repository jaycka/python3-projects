A low-risk major is a degree where there is a small difference between the lowest and highest salaries. In other words, if the difference between the 10th percentile and the 90th percentile earnings of your major is small, then you can be more certain about your salary after you graduate.

```python
spread = clean_df['Mid-Career 90th Percentile Salary']-clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, 'Spread',spread)
```

```python
clean_df.sort_values('Spread', ascending=True).head(10)
```

Top 5 Majors with the Highest Potential

```python
clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)[['Undergraduate Major','Mid-Career 90th Percentile Salary']].head()
```

Top 5 Majors with the Greatest Spread in Salaries

```python
clean_df.sort_values('Spread', ascending=False)[['Undergraduate Major','Spread']].head()
```

What is the average salary by group?

```python
pd.options.display.float_format='{:,.2f}'.format
clean_df.groupby('Group').mean()
```
